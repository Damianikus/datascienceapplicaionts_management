```python
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
```

```python
def clean_stopwords(text: str, language:str = "english"):
    words = text.split(" ")
    for word in words:
        if word not in stopwords.words(language):
            print(word)
```

```python
text = "Hallo Welt wie  geht es dir?"
```

```python
clean_stopwords(pipeline(text), "german")
```
